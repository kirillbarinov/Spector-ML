```python
from pathlib import Path
import pandas as pd

# Define the directory where the files are located
input_files_dir = Path('zd/')

# List all Reflection files in the directory
reflection_files = list(input_files_dir.glob('Reflection__*.txt'))
print(len(reflection_files))

# Initialize an empty DataFrame to store all measurements
all_measurements_df = pd.DataFrame()
all_measurements_df.shape
o = 0
# Process each file
for file_path in reflection_files:
    # Open the input file and read its content
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Find the start of the spectral data
    start_index = 0
    for i, line in enumerate(lines):
        if '>>>>>Begin Spectral Data<<<<<' in line:
            start_index = i + 1
            break

    # Extract only the measurements to a list
    measurements = [line.strip().split('\t')[1] for line in lines[start_index:] if line.strip()]

    # Create a DataFrame for the current file's measurements and transpose it
    df_measurements = pd.DataFrame(measurements).T

    # Append the current file's measurements to the all_measurements_df
    all_measurements_df = pd.concat([all_measurements_df, df_measurements], ignore_index=True)

# Reset the index of the final DataFrame
all_measurements_df.reset_index(drop=True, inplace=True)

# Display the shape of the DataFrame to verify the concatenation process
all_measurements_df.shape
all_measurements_df
```

```python
all_measurements_df.to_excel('zdorov.xlsx')
```
